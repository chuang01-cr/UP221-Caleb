## Creating an Index of Transit Vulnerability in San Francisco
To understand which areas are in most need of transit access to high schools, we are creating an index. The index considers a census tract's percentage of households living under the poverty line, the percentage of people who take transit to work, the percentage of people who are non-White, and the percentage of people who are aged 5-17. We will then observe the presence of high schools and transit routes that connect to a high school to determine whether there is access to education from that neighborhood. 

```python
import geopandas as gpd
import matplotlib.pyplot as plt 
import pandas as pd
import mapclassify
import folium
import contextily as ctx
```

```python
tracts = gpd.read_file('data/SF Census Tracts.geojson')
df = pd.read_csv('data/census.csv')
tracts.head()
```

```python
tracts['GEOID'] = tracts['geoid'].astype(str)
tracts.head()
```

```python
df['GEOID'] = df['GEOID'].astype(str)
df['GEOID'] = df['GEOID'].str.zfill(11)
df.head()
```

```python
merged_tracts = tracts.merge(df, how='left', on='geoid')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Set fig, ax and plot with legend outside of plot
fig, ax = plt.subplots(figsize=(10, 10))
merged_tracts.plot(column="Percentage of people who ride transit", scheme="Natural_Breaks", k=5, cmap="YlGn", linewidth=0, legend=True, ax=ax)

# Use tight layout
plt.tight_layout()
plt.title('Percentage of Non-White Residents in LA County')

ax.axis('off')

# Place legend outside of plot
ax.get_legend().set_bbox_to_anchor((1.5, 1))
```

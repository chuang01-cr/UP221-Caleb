```python
total_count_columns = ['Geo_FIPS',
                       'SE_A04001_001',
                       'SE_A04001_002',
                       'SE_A04001_003',
                       'SE_A04001_004',
                       'SE_A04001_005',
                       'SE_A04001_006',
                       'SE_A04001_007',
                       'SE_A04001_008',
                       'SE_A04001_009',
                      'SE_A04001_010']

# creating a new dataframe with the desired columns
df2 = df[total_count_columns]
```

```python
# read census tracts datafile
tracts=gpd.read_file('Data/SF Census Tracts.geojson')
tracts.head()
```

```python
tracts.plot(figsize=(12,10))
```

```python
# looking at the tract's columns
tracts.info()
```

```python
# filtering for the columns I want to see
tracts = tracts[['namelsad', 'tractce', 'geoid', 'geometry']]
tracts.head()
```

```python
# renaming GEOID to FIPS
tracts = tracts.rename(columns={'geoid': 'FIPS'})
tracts.head()
```

```python
# creating a new dataframe so I can merge the two data sets
tracts_race = tracts.merge(df2,on="FIPS")
tracts_race.head()
```

```python
# plotting a map showing concentration of Asian population in San Francisco
tracts_race.plot(figsize=(12,10),
                 column='NH Asian Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
# Finding all census tracts with no people again
tracts_race[tracts_race['Total Pop']==0]
```

```python
# deleting all census tracts with no people
tracts_race.drop([35,88,205],axis=0,inplace=True)
```

```python
# plotting a new map
tracts_race.plot(figsize=(12,10),
                 column='NH Asian Alone',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
# Creating new columns to show percentages
tracts_race['PCT_Black'] = tracts_race['NH Black Alone']/tracts_race['Total Pop']*100
tracts_race['PCT_White'] = tracts_race['NH White Alone']/tracts_race['Total Pop']*100
tracts_race['PCT_Asian'] = tracts_race['NH Asian Alone']/tracts_race['Total Pop']*100
tracts_race['PCT_Hispanic'] = tracts_race['Hispanic or Latino']/tracts_race['Total Pop']*100
```

```python
# plotting a map showing concentration of Black residents in San Francisco
tracts_race.plot(figsize=(12,10),
                 column='PCT_Black',
                 legend=True, 
                 scheme='equal_interval')
```

```python
# plotting a map showing concentration of Asian residents in San Francisco
tracts_race.plot(figsize=(12,10),
                 column='PCT_Asian',
                 legend=True, 
                 scheme='equal_interval')
```

```python
# plotting a map showing concentration of White residents in San Francisco
tracts_race.plot(figsize=(12,10),
                 column='PCT_White',
                 legend=True, 
                 scheme='equal_interval')
```

```python
# plotting a map showing concentration of Hispanic or Latino residents in San Francisco
tracts_race.plot(figsize=(12,10),
                 column='PCT_Hispanic',
                 legend=True, 
                 scheme='equal_interval')
```

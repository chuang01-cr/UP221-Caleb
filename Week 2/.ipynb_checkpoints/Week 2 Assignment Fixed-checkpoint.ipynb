```python
# import pandas and geopandas for map creating
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
```

```python
# importing the data file
muni = gpd.read_file('Data/Muni Stops_20250119.zip')
```

```python
# observing the fist 5 rows
muni.head()
```

```python
# observing street counts
Street_Stop_Count = muni['onstreet'].value_counts()
Street_Stop_Count
```

```python
# retrieve info about the data file's columns
muni.info()
```

```python
# understand how many rows (number of stops) and columns
print(muni.shape)
```

```python
# create list of wanted columns
desired_columns = ['stopname', 'latitude', 'longitude', 'onstreet', 'geometry']

# testing to see if it worked
muni[desired_columns].sample(5)
```

```python
# filter out unwanted columns and save the variable
muni_trimmed = muni[desired_columns].copy()
muni_trimmed
```

```python
# checking my work
muni_trimmed.info()
```

```python
muni_trimmed.plot()
```

```python
# filtering to see which stops have the on street of Broad
Broad_stops = muni_trimmed[muni_trimmed.onstreet == 'BROAD ST']
Broad_stops
```

```python
# plotting Broad St stops
Broad_stops.plot()
```

```python
# creating a bigger map
muni_trimmed.plot(
            figsize=(20,12),   #size of the plot (a bit bigger than the default
)
```
